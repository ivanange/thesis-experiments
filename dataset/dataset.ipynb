```python
import pandas as pd
import numpy as np
```

```python
#load image_scores with columns file_path, adult_score, racy_score, mean_score, category, class
image_scores = pd.read_csv('image_scores.csv')
random_state = 42
# set random seed
np.random.seed(random_state)
```

```python
# add column score with maximum of adult_score and racy_score
image_scores['score'] = image_scores[['adult_score', 'racy_score']].max(axis=1)

invalid_images = image_scores[image_scores['category'] == 'invalid']
invalid_images = invalid_images.sort_values(by='score', ascending=False)
len(invalid_images)
```

```python
# create empty dataframe with columns file_path, score, category, class
dataset = pd.DataFrame(columns=['file_path', 'score', 'category', 'time', 'censored', 'class'])

sorted_invalid_images = invalid_images.copy()

# while there are still invalid images pick i + k images with highest score
i = 0
k = 1
while len(sorted_invalid_images) > 0:
    k = j = min(np.random.randint(1, 5), len(sorted_invalid_images))
    for index, row in sorted_invalid_images.iterrows():
        if j > 0:
            dataset = dataset.append({
                'file_path': row['file_path'],
                'score': row['score'],
                'category': row['category'],
                'time': i+1,
                'censored': 1,
                'class': row['class']
            }, ignore_index=True)
            j -= 1
        else:
            break
    
    # remove i + k images with highest score
    sorted_invalid_images = sorted_invalid_images.iloc[k:]
    i += 1
    
dataset
```

```python
# group by time
grouped_dataset = dataset.groupby('time')
grouped_dataset.describe()
```

```python
# add valid samples to dataset
valid_images = image_scores[image_scores['category'] == 'valid'].drop(columns=['adult_score', 'racy_score', 'mean_score',])
valid_images['time'] = i+1
valid_images['censored'] = 0
dataset = dataset.append(valid_images, ignore_index=True)
# for index, row in valid_images.iterrows():
#     dataset = dataset.append({
#         'file_path': row['file_path'],
#         'score': row['score'],
#         'category': row['category'],
#         'time': i,
#         'censored': 0,
#         'class': row['class']
#     }, ignore_index=True)
#     j -= 1
dataset
```

```python
dataset.to_csv('dataset.csv', index=False)
```
